```python
#use legend
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
months = [1,2,3,4]
sales = [1100,1300,1500,1200]
plt.plot(months,sales,color ='blue',linestyle = '-',linewidth =2, marker = 'o',label = '2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales of year')
plt.legend(loc ='upper left',fontsize = 15,)
plt.show()
```

```python



#plt.grid(color ='colorname',linestyle=' ',linewidth='width')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
months = [1,2,3,4]
sales = [1100,1300,1500,1200]
plt.plot(months,sales,color ='blue',linestyle = '-',linewidth =2, marker = 'o',label = '2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales of year')
plt.legend(loc = 'lower right',fontsize =10)
plt.grid(color ='gray',linestyle=':',linewidth=2)
plt.xlim(1,4)
plt.ylim(0,1000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```
