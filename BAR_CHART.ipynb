```python
import matplotlib.pyplot as plt
```

```python
#plt.bar(x,heigh,color='colornsme',width = 'value',label ='lable name')
items = ["A","B","C","D","E"]
sales = [1000,1200,1500,800,1800]
plt.bar(items,sales,color = "green",width = 0.8,label = "sales 2025")
plt.xlabel("Items")
plt.ylabel("Sales")
plt.show()
```
