```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

# LINE PLOT
plt.subplot(1, 2, 1)
plt.plot([1,2,3,4], [5,6,2,8])
plt.title("LINE PLOT")

# BAR CHART
plt.subplot(1, 2, 2)
plt.bar([1,2,3,4], [5,3,6,2])
plt.title("BAR CHART")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

# Line plot
plt.subplot(2, 2, 1)
plt.plot([1,2,3], [3,4,5])
plt.title("LINE PLOT")

#scatter plot
plt.subplot(2, 2, 2)
plt.scatter([1,2,3], [5,1,3])
plt.title("SCATTER PLOT")

#bar chart
plt.subplot(2, 2, 3)
plt.bar([1,2,3], [4,2,5])
plt.title("BAR CHART")

#histogram
plt.subplot(2, 2, 4)
plt.hist([1,1,2,3,3,3,4])
plt.title("HISTOGRAM")

plt.tight_layout()
plt.show()
```
