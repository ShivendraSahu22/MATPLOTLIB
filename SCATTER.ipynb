```python
import matplotlib.pyplot as plt
```

```python

X = [1, 5, 6, 7, 77, 88, 55, 65]
Y = [4, 3, 6, 7, 4, 55, 34, 3]
plt.scatter(X, Y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot Example")
plt.show()
```
